# **Playing with matplotlib**

```python
import numpy as np
import matplotlib.pyplot as plt
```

## **Scatter plot**

```python
X_data = np.random.random(size = 50)
y_data = np.random.random(size = 50)
```

### **Basic plot**

```python
plt.scatter(X_data, y_data);
```

### **Playing with options**

```python
plt.scatter(x = X_data, y = y_data, color = '#0000ff', marker = '*', s = 42);
```

## **Line plot**

```python
years = [2006 + x for x in range(15)]
print(years)
```

```python
weights = [100 + y for y in np.random.randint(low = -5, high = 5, size = 15).tolist()]
print(weights)
```

### **Basic plot**

```python
plt.plot(years, weights);
```

### **Playing with options**

```python
plt.plot(years, weights, color = '#ff8800', linewidth = 3, linestyle = '--');
```

```python
weights_2 = [100 + y for y in np.random.randint(low = -8, high = 8, size = 15).tolist()]
print(weights_2)
```

```python
plt.plot(years, weights, color = '#ff8800', linewidth = 3, linestyle = '--')
plt.plot(years, weights_2, color = '#88ff00', linewidth = 3, linestyle = ':');
```

## **Bar chart plot**

### **Basic plot**

```python
prog_lang = ['C++', 'C#', 'Python', 'Java', 'Go']
votes = [45, 55, 75, 50, 35]
```

```python
plt.bar(x = prog_lang, height = votes);
```

### **Playing with options**

```python
plt.bar(x = prog_lang, height = votes, color = '#994400', edgecolor = '#000000', width = 0.5, linewidth = 3);
```

## **Histogram**

```python
hist_data = np.random.normal(loc = 20, scale = 1.5, size = 1000)
```

### **Basic plot**

```python
plt.hist(x = hist_data);
```

### **Playing with options**

```python
plt.hist(x = hist_data, bins = 50);
```

```python
plt.hist(x = hist_data, bins = [hist_data.min(), 16, 18, hist_data.mean(), 22, 24, hist_data.max()], cumulative = True);
```

## **Piechart**

### **Basic plot**

```python
plt.pie(x = votes, labels = prog_lang);
```

### **Playing with options**

```python
explodes = [0.05] * len(prog_lang)
plt.pie(x = votes, labels = prog_lang, explode = explodes, autopct = '%.2f%%', startangle = 90);
```

## **Customizing the graph**

### **Changing the font**

```python
years = [x for x in range(2016, 2026)]
income = [(x + x * round(np.random.random(), 2)) for x in range(50, 60)]
income_ticks = [x for x in range(55, 101, 5)]
```

```python
fontdict = {
  'family': 'serif',
  'color':  'darkred',
  'weight': 'normal',
  'size': 16,
}
```

```python
plt.plot(years, income)
plt.title('Income of John Doe (in USD)', fontsize = 24)
plt.xlabel('Year', fontdict = fontdict)
plt.ylabel('Yearly income (in USD)', fontdict = fontdict)
plt.xticks(years)
plt.yticks(income_ticks, [f'{x}k USD' for x in income_ticks]);
```

### **Adding a legend**

```python
income_b = [(x + x * round(np.random.random(), 2)) for x in range(50, 60)]
income_c = [(x + x * round(np.random.random(), 2)) for x in range(50, 60)]
```

```python
income_ticks_2 = [x for x in range (50, 125, 5)]
```

```python
plt.plot(years, income, label = 'Income of John Doe')
plt.plot(years, income_b, label = 'Income of Jane Doe')
plt.plot(years, income_c, label = 'Income of Jack Doe');
```

```python
plt.figure(figsize = (10, 5))
plt.plot(years, income, label = 'Income of John Doe')
plt.plot(years, income_b, label = 'Income of Jane Doe')
plt.plot(years, income_c, label = 'Income of Jack Doe')
plt.xticks(years)
plt.yticks(income_ticks_2)
plt.legend(loc = 'upper left');
```
